```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
import numpy as np
```

### Storing 3 tables present in the databse  into 3 dataframes 

```python
db = sqlite3.connect('IMDB.sqlite')
db
```

```python
cur = db.cursor()
cur
```

### IMDB dataframe

```python
imdb = pd.read_sql_query('select * from imdb' , db)
imdb
```

### earning dataframe

```python
earn = pd.read_sql_query('select * from earning' , db)
earn
```

### genre dataframe

```python
genre_df = pd.read_sql_query('select * from genre' , db)
genre_df
```

### 1. Genre and number of movies present in that genre.

```python
genre_mod_df = genre_df[genre_df['genre'] != '']
genre_mod_df.to_sql("genre_mod", db , if_exists = "replace")
cur.execute('select genre , count(*) from genre_mod group by genre')
for row in cur:
    l = [row[0] , row[1]]
    print(l[0] , l[1])
```

### 2. Rating of the highest-grossing movie worldwide.

```python
earn['total_earning'] = earn['Domestic'] + earn['Worldwide']

index = earn.total_earning.idxmax()
movie_id = earn['Movie_id'].iloc[index]
loc  = imdb.Rating.loc[imdb['Movie_id'] == movie_id]

print(loc.values[0])
```

### 3 . Movie name and rating of the most voted movie.

```python
index = imdb.TotalVotes.idxmax()
r = imdb['Rating'].iloc[index]
t = imdb['Title'].iloc[index]
print(t,r)
```

### 4 . Movie name and running time of the movie with Longest-Running Time.

```python
def getTime(time):
    return int(time.split(' ')[0])
imdb['Runtime'].replace('','0',inplace=True) 
imdb['Runtime']=imdb['Runtime'].apply(getTime)

index = imdb.Runtime.idxmax()
n = imdb.Title.iloc[index]
t = imdb.Runtime.max()

print(n,t)
```

### 5 . Movie with maximum net profit .

```python
df = pd.merge(imdb , earn , on = 'Movie_id')
df['Domestic'].replace('',0,inplace = True)
df['Worldwide'].replace('',0,inplace = True)

def f(val):
    return int(val)
df['Budget'].replace('','0',inplace=True) 
df['Budget']=df['Budget'].apply(f)



df['profit'] = df['Domestic'] + df['Worldwide'] - df['Budget']

index = df.profit.idxmax()
print(df.Title.iloc[index])
```

### 6 . Least budgeted movie with maximum net profit .

```python
df = pd.merge(imdb , earn , on = 'Movie_id')
df['Domestic'].replace('',0,inplace = True)
df['Worldwide'].replace('',0,inplace = True)

def f(val):
    return int(val)
df['Budget'].replace('','0',inplace=True) 
df['Budget']=df['Budget'].apply(f)



df = df[df['Budget'] != 0]
df = df[df['Budget'] == df['Budget'].min()]
df['profit'] = df['Domestic'] + df['Worldwide'] - df['Budget']

ind = df.profit.idxmax()
print(df.Title.loc[ind])
```

### 7. Highest grossing movies yearwise.

```python
data=pd.read_sql_query('select imdb.title, earning.domestic, earning.worldwide from imdb join earning on imdb.movie_id=earning.movie_id;', db)
df=data.copy()
df['Gross']=df.Domestic+df.Worldwide
name=[]
year=[]
for i in df.Title:
    year.append(i[len(i)-5:len(i)-1])
    name.append(i[:len(i)-6].strip())
np_name=np.array(name)
np_year=np.array(year, dtype='int')
df['Year']=np_year
df['Name']=np_name
list_of_years=list(set(np_year))
list_of_years=np.sort(list_of_years)
for i in list_of_years:
    print(i, df.Title[np.argmax(df[df.Year==i].Gross)])
```

### 8 . Percent of budget for each genre in the form of a pie chart.

```python
data = pd.read_sql_query('select imdb.Budget , genre.genre from imdb join genre on imdb.Movie_id = genre.Movie_id' , db)
df = data.copy()
df = df[df.genre != '']
df.Budget.replace('',0,inplace = True)
gen_bud = df.groupby('genre')['Budget'].sum().sort_values(ascending = False)
gen = gen_bud.index;
bud = gen_bud.values
plt.rcParams['figure.figsize'] = (15,9)
plt.pie(bud,labels = gen,autopct='%.2f%%' , startangle = 90 ,shadow = True,wedgeprops   = { 'linewidth' : 2,'edgecolor' : "orange" })
plt.show()
per = np.true_divide(bud , bud.sum())*100
for i in range(len(per)):
    print(gen[i] , format(per[i] , '.2f'))
```

### 9 . Percentage of total earning for each genre.

```python
data = pd.read_sql_query('select earning.Domestic , earning.Worldwide , genre.genre from earning join genre on earning.Movie_id = genre.Movie_id' , db)
df = data.copy()
df = df[df.genre != '']
df.Domestic.replace('',0,inplace = True)
df.Worldwide.replace('',0,inplace = True)
df['total'] = df['Domestic'] + df['Worldwide']
gen_bud = df.groupby('genre')['total'].sum().sort_values(ascending = False)
gen = gen_bud.index;
bud = gen_bud.values
plt.rcParams['figure.figsize'] = (15,9)
plt.pie(bud,labels = gen,autopct='%.2f%%' , startangle = 90 ,shadow = True,wedgeprops   = { 'linewidth' : 2,'edgecolor' : "cyan" })
plt.show()
per = np.true_divide(bud , bud.sum())*100
for i in range(len(per)):
    print(gen[i] , format(per[i] , '.2f'))
```
